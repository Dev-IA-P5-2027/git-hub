```python
# Écrire une fonction qui convertit en mètres par seconde et en km/h une vitesse fournie en paramètres par l'utilisateur en miles/heure (1 mile = 1609 mètres). Ce résultat sera retourné par la fonction.

def Vitesse():
    Miles = int(input("entre ta vitesse en mille/h"))
    KM = Miles*1.609
    Ms = KM/3.6
    print(f"La vitesse du véhicule est de {KM} km/h et de {Ms} m/s")

Vitesse()
```

```python
# Écrire un programme en python pour réviser ses tables de multiplication. Le programme tire au hasard deux entiers entre 0 et 10 et demande à l'utilisateur de donner leur produit.
import random

def Table_de_multiplication():
    Pts = 0
    for i in range(5):
        # A = math.random(0,10)
        A = random.randint(1,10)
        B = random.randint(1,10)
        print(f"{A}*{B}=")
        Réponse = int(input("entre ta réponse"))
        if Réponse == A*B:
            Pts+=1 
            print("Tu gagne un point")
            print(f"Tu as {Pts}")
        else:
            Pts-=1
            print("Tu perd un point")
            print(f"Tu as {Pts}")
        if Pts == 5:
            print("Tu as gagné")
        else:
            print("Suivant")

Table_de_multiplication()
```
